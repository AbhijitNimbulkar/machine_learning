```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)

#n_samples: This parameter determines the number of samples to be generated in the dataset. In your case, you're generating 100 samples.
#n_features: This parameter specifies the number of features (independent variables) each sample will have. Here, you're generating a dataset with 2 features per sample.
#n_informative: This parameter specifies the number of informative features, meaning the number of features that actually influence the target variable. In your case, you've set it to 2, indicating that 2 out of the 2 features are informative.
#n_targets: This parameter specifies the number of target variables (dependent variables) in the dataset. You've set it to 1, meaning you're generating data for a single target variable.
#noise: This parameter determines the standard deviation of the gaussian noise applied to the output. It represents the amount of random variation or "noise" in the target variable not accounted for by the features. In your case, you've set it to 50, indicating a relatively high level of noise.

#Here's a way to understand it:

#Let's say the actual price of a house is $200,000.
#If we set noise=50, it means that when we generate fake data for this house, the price could be anywhere between $150,000 and $250,000. This range is determined by adding or subtracting 50 (the standard deviation) from the actual price.
#So, the larger the value of noise, the more spread out the generated prices will be around the actual prices.
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generating synthetic data with make_regression
X, y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)

# Checking the shapes of X and y
print("Shape of X:", X.shape)  # Should print (100, 2) if X has 100 samples and 2 features
print("Shape of y:", y.shape)  # Should print (100,) if y has 100 samples

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)

# Checking the shapes of the training and testing sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)



#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_regression

# Generate synthetic regression data
X, y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)

# Assuming x, y, and z represent the grid for your features and predicted values
# For this example, let's assume x and y are grids created from the feature data
x = np.linspace(min(X[:, 0]), max(X[:, 0]), 10)
y = np.linspace(min(X[:, 1]), max(X[:, 1]), 10)
x, y = np.meshgrid(x, y)  # Create grid from x and y

# Assuming z represents the predicted target variable values
# For this example, let's assume z is generated by a simple linear model
z = np.dot(np.column_stack((x.flatten(), y.flatten())), np.random.rand(2)) + np.random.rand(100)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the data points
ax.scatter(X[:, 0], X[:, 1], y, c='blue', marker='o')

# Surface plot for the predicted values
ax.plot_surface(x, y, z.reshape(x.shape), color='red', alpha=0.5)

# Set labels
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')

# Show the plot
plt.show()


```
