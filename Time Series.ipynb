```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
date_range = pd.date_range(start='2022-01-01' , end='2023-12-31',freq='D')
level=100
trend = np.linspace(0,10,len(date_range))
seasonality = 10*np.sin(2 *np.pi * date_range.dayofyear / 365)
noise = np.random.normal(0,2,len(date_range))
```

```python
data = level + trend + seasonality + noise
time_series = pd.DataFrame({'Date': date_range,'Value':data})
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date' ,y='Value',data=time_series,label='Time-Series Data')
plt.title('Time-Series Data with Level,Trend,and Seasonality')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
