# Bayes Theorem Derivation
Bayes Theorem can be derived for events and random variables separately using the definition of conditional probability and density.

Formula : P(A|B) = P(B|A)P(A) / P(B)

where,

P(A) and P(B) are the probabilities of events A and B

P(A|B) is the probability of event A when event B happens

P(B|A) is the probability of event B when A happens

# Types of Naive Bayes


1.Bernoulli Naive Bayes : Assumes that  feature are BInary  (Boolean ) Variable (o,1).

2.Gaussion Naive Bayes : Assumes that Continuous feature follow a Gaussion (Normal ) Distribution.

3.Multinominal Naive Bayes : Used for Discrete Data , Text Data  That Feature Follow Multinominal Distribution.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placement.csv")
dataset
```

```python
dataset.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
dataset.shape
```

```python
dataset.shape[0]
```

```python
dataset.isnull().sum().sum()
```

```python
sns.scatterplot(x="cgpa", y="Score", data = dataset , hue ="placement" )
plt.legend(loc = 2)
plt.show()
```

```python
x= dataset.iloc[:,:-1]
y= dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=dataset["Score"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

# decision_regions (GaussianNB)

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() , clf =gnb)
plt.show()
```

```python
mnb= MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100
```

# Decision Regions(MultinomialNB)

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() , clf =mnb)
plt.show()
```

```python
bnb =BernoulliNB()
bnb.fit(x_test,y_test)
```

```python
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
```

# Decision Regions(BernoulliNB)

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() , clf =bnb)
plt.show()
```
